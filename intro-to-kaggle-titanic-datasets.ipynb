```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
#dataset sample columns
train_data.head()
```

```python
#basic statistics of the training dataset
train_data.describe()
```

```python
# test data description
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.describe()
```

```python
women = train_data[train_data['Sex'] == 'female']['Survived']
rate_women_survived = sum(women)/ len(women)
print(" The percentage of women who has survived is {}".format( rate_women_survived))

men = train_data[train_data['Sex'] == 'male']['Survived']
rate_men_survived = sum(men)/ len(women)
print(" The percentage of women who has survived is {}".format( rate_men_survived))
```

```python
#Machine learning algorithm application
#import machine learning library
import sklearn.ensemble 
from sklearn.ensemble import RandomForestClassifier
```

```python
y_tr = train_data["Survived"]

#Relevant features
features = ["Pclass", "Sex", "SibSp", "Parch"]

#converting string/categorical variables into dummy/indicator variables
X_tr = pd.get_dummies( train_data[ features ] )
X_ts = pd.get_dummies( test_data [ features ] )


classifier = RandomForestClassifier( n_estimators = 100, max_depth = 5, random_state= 1)
classifier.fit(X_tr,y_tr)

#survival predictions
pred = classifier.predict( X_ts )

print( pred[:5] )

output = pd.DataFrame({ 'PassengerId': test_data['PassengerId'], 'Survived': pred })
output.to_csv("my_submission.csv", index=False)
print("Your submission was successfully saved!")
```
